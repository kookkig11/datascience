## 13

13. สมมติว่าเรานิยามแผ่นดินไหวที่รุนแรงนั้นต้องมี Magnitude มากกว่าการไหวครั้งก่อนและครั้งหลังอย่าง
น้อย 2.0 จงคัดเลือกเหตุการณ์แผ่นดินไหวที่รุนแรงออกมา

```python

newData["DiffBefore"] = dataset.Magnitude.diff()
newData["DiffAfter"] = dataset.Magnitude - dataset.Magnitude.shift(-1)
newData.loc[(newData.DiffBefore >= 2) & (newData.DiffAfter >= 2),:]
```

```python
impactMag = newData.loc[(newData.DiffBefore >= 2) & (newData.DiffAfter >= 2),:]
impactMag.index.date
impactMag.index.hour
```
