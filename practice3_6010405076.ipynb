#### 6.	แผ่นดินไหวที่แรงที่สุดกับแรงเป็นที่ 2 ห่างกันเท่าไหร่ ตอบเป็นวัน-ชั่วโมง-นาที

```python
maxMag = eq.loc[eq['Magnitude'] == eq['Magnitude'].max()]
maxMag
```

```python
# ascending=False >> max..min
secMag = eq.sort_values('Magnitude', ascending=False)[1:2]
secMag
```

```python
# secMag before maxMag
# second > hour > minute
day = (maxMag.index - secMag.index)[0].days
s = (maxMag.index - secMag.index)[0].seconds
h = s//3600
m = s//60 - (h*60)
print("%i day - %i hour - %i minute" %(day, h, m))
```

#### 11.	ระยะเวลาที่ไม่มีแผ่นดินไหวเลยนานที่สุดคือเท่าไหร่

#### 12.	ระยะเวลาที่ไม่มีแผ่นดินไหวเลยนานที่สุดคือช่วงไหน ระบุเป็นเวลาเร่ิมต้นและเวลาจบ

```python
# ระยะเวลา difTime
difTime = newData.diff()
rangeTD = difTime.loc[difTime.Range == difTime.Range.max()].Range
rangeTD
```

```python
# sub with rangeTD = start
start = (difTime.loc[difTime.Range == difTime.Range.max()].index - rangeTD)[0]
start
```

```python
end = difTime.loc[difTime.Range == difTime.Range.max()].index[0]
end
```

```python
print("Start :", start.strftime('%Y-%m-%d %H:%M.%S'),
      "\nStop : ",end.strftime('%Y-%m-%d %H:%M.%S'))
```

#### 13.	สมมติว่าเรานิยามแผ่นดินไหวที่รุนแรงนั้นต้องมี Magnitude มากกว่าการไหวครั้งก่อนและครั้งหลังอย่างน้อย 2.0
#### จงคัดเลือกเหตุการณ์แผ่นดินไหวที่รุนแรงออกมา

```python
newData["DiffBefore"] = eq.Magnitude.diff()
newData["DiffAfter"] = eq.Magnitude - eq.Magnitude.shift(-1)  # shift เลื่อนข้อมูล : -1 index ก่อนหน้า
newData.loc[(newData.DiffBefore >= 2) & (newData.DiffAfter >= 2),:]
```

```python
impactMag = newData.loc[(newData.DiffBefore >= 2) & (newData.DiffAfter >= 2),:]
impactMag.index.date
impactMag.index.hour
```

#### 14.	จากตารางข้อ 13. ให้เพิ่มคอลัมน์ใหม่ชื่อ hour_count แสดงจำนวนแผ่นดินไหวในชั่วโมงที่มีเหตุการณ์ แผ่นดินไหวรุนแรงด้วย

```python
# มี impactMag.index.date ใน eq.index.date
eqCombineMag = eq.loc[(np.isin(eq.index.date, impactMag.index.date))
                    & (np.isin(eq.index.hour, impactMag.index.hour))]
eqCombineMag
```

```python
hour_count = eqCombineMag.groupby([eqCombineMag.index.date, eqCombineMag.index.hour]).size()
hour_count
```

#### 15.	แสดงจำนวนแผ่นดินไหวแยกตาม 24 ชั่วโมง (เอาจำนวนแผ่นดินไหวของแต่ละชั่วโมงของทุกวันรวมกัน แสดงผลตามชั่วโมง)

```python
eq["PerHour"] = 0
eq.PerHour = eq.loc[eq.Magnitude != 0].PerHour = 1
eq
```

```python
alltime = eq.resample("1h").sum()
alltime
```

```python
days = pd.date_range("2016-08-24", "2016-11-30", freq='1H')
days = pd.DataFrame(days)
#days = days.set_index(0)
days.columns = ["Time"]
days
```

```python
result = pd.merge(days, alltime, how='left', on=["Time"])
result = result.fillna(0)
result = result.set_index("Time")
recentlyResult = result.groupby([result.index.date, result.index.hour]).sum()
recentlyResult.loc[:,["PerHour"]]
```
