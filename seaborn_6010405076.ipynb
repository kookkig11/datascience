## 1.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("players.csv")
df = df.dropna(axis='columns') # drop NaN columns
df
```

```python
tmp = df.loc[:,['shots', 'passes', 'tackles', 'saves']]
perMinute = tmp.div(df.minutes, axis=0)
perMinute
```

```python
df = df.drop(columns=['shots', 'passes', 'tackles', 'saves'])
df = df.join(perMinute)
df
```

```python
df.position = df.position.astype('category')
df.position = pd.Categorical(df.position, categories=['goalkeeper', 'defender', 'midfielder', 'forward'], ordered=True)
df
```

# 2.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(df.passes)
```

## 3.

```python
USA = df.loc[df.team.str.contains('USA'),:]
```

```python
sns.distplot(USA.shots, kde=False, rug=True)
```

## 4.

```python
sns.relplot(data=df, x='passes', y='shots')
```

## 5.

```python
teamBSU = df.loc[df.team.str.contains('Brazil') |
                 df.team.str.contains('Spain') |
                 df.team.str.contains('USA')]
teamBSU
```

```python
# color:hue สัญลักษณ์:style
sns.relplot(data=teamBSU, x='passes', y='shots', hue='team', style='position')
```

## 6.

```python
sns.relplot(data=teamBSU, x='passes', y='shots', col='team', row='position')
```

## 7.

```python
sns.catplot(data=teamBSU, x='team', y='passes', kind='box')
```

## 8.

```python
sns.catplot(data=teamBSU, x='team', y='passes', kind='swarm', hue='position')
```

## 9.

```python
sns.catplot(data=teamBSU, x='position', y='tackles', kind='bar', hue='team')
```
