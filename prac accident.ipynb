#### 9.	วันที่เกิดเหตุนั้นมีตั้งแต่ 27 (ธันวาคม) จนถึง 5 (มกราคม) แต่การตั้งค่าวันแบบนี้จะทำให้การเรียงลำดับวัน ผิดพลาด ให้แก้วันของเดือนธันวาคมให้เป็นค่าติดลบย้อนหลัง เช่น 31 จะเป็น -1 วันที่ 30 จะเป็น -2 ไป เรื่อยๆ วันของเดือนมกราคมให้คงเดิม (1-5) ส่วนวันที่เป็น 0 ให้ทิ้งแถวนั้นไป

```python
acc.loc[acc['date'] == 31, 'date'] = -1
acc.loc[acc['date'] == 30, 'date'] = -2
acc.loc[acc['date'] == 29, 'date'] = -3
acc.loc[acc['date'] == 28, 'date'] = -4
acc.loc[acc['date'] == 27, 'date'] = -5
acc = acc.loc[acc['date'] != 0]
acc
```

```python
def toFloat(x):
    if type(x) != float:
        x = x.replace("[","")
        x = x.replace("]","")
        x = x.replace(",","")
    else:
        x = float(x)
    return x
withDis58["Distance"].apply(toFloat).astype(np.float64)
```
