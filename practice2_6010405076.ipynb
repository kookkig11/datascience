#### 3. drinks มีข้อมูลที่คอลัมน์ไหนที่เป็น NaN บ้าง และแต่ NaN กี่ตัวในแต่ละคอลัมน์

#### 4. ลองเช็คว่าชื่อประเทศของตาราง medals ใดที่ไม่มีอยู่ในตาราง drinks
#### 1.การที่จะ join ตารางได้นั้นต้องมีคอลัมน์ที่มีข้อมูลเหมือนกันในทั้งสองตาราง ในที่นี้ให้ใช้ชื่อประเทศ
#### 2.สกัดเอาชื่อประเทศจากตาราง medals ออกมาโดยใช้คำสั่ง str.extract("(.+)\s\\(") มาสร้างเป็นคอ ลัมน์ใหม่
#### 3.ลองใช้คำสั่ง isin() ในการเช็คว่ามีประเทศใดใน medals ไม่อยู่ใน drinks บ้าง

```python
medals['country'] = medals['country'].str.extract("(.+)\s\(")
medals
```

```python
# ~medals['country'] ไม่มีใน medals['country']
medals.loc[~medals['country'].isin(drinks['country']) == True, ['country']]
```

#### 5. แสดงชื่อประเทศของตาราง drinks ที่ไม่มีอยู่ในตาราง medals
#### 1.	นิสิตจะเห็นความแตกต่างของชื่อประเทศในสองตารางนี้ว่าเรียกชื่อประเทศไม่เหมือนกัน ตอนทำงาน จริงนั้นต้องปรับชื่อให้ตรงกัน (เช่น USA กับ United States) แต่ในการบ้านนี้ไม่ต้องทำ

```python
drinks.loc[~drinks['country'].isin(medals['country']) == True, ['country']]
```

#### 6. นำคอลัมน์ที่ชื่อ continent ของ drinks ไปต่อกับตาราง medals ตามชื่อประเทศของ medals

```python
medals['continent'] = medals.merge(drinks, on='country').continent
medals
```

#### 7. ในตาราง medals จงแสดงชื่อประเทศที่ยังไม่มีข้อมูลในคอลัมน์ continent

```python
medals.loc[ medals['continent'].isnull(), ['country']]
```

#### 8. นิสิตจะเห็นแนวทางการแก้ไขข้อมูลหรือรวมประเทศบางประเทศเข้าด้วยกัน แต่ในการบ้านนี้ยังไม่ต้องทำ แค่เติมข้อมูลทวีปที่ยังขาดหายไปในบางประเทศดังนี้
#### Australasia: OC, Canada/United States/ Mexico: NA, Cuba/Jamaica: SA, Czech/Yugoslavia/Germany(ทั้งหมด)/Great Britain/Russia(ทั้งหมด)/Soviet: EU ประเทศที่เหลือกำหนดให้เป็น 'OT' ไป
#### 1.	ตารางนี้จะเป็นตารางแบบ wide ที่สมบูรณ์ ควรเก็บไว้เพื่อใช้ในการคำนวณ

```python
medals.loc[ medals['country'].str.contains('Australasia'), 'continent' ] = "OC"
medals.loc[ medals['country'].str.contains('Canad') | medals['country'].str.contains('United S')
          | medals['country'].str.contains('Mexico'), 'continent' ] = "NA"
medals.loc[ medals['country'].str.contains('Ger') | medals['country'].str.contains('Czech') |
            medals['country'].str.contains('Yugo') | medals['country'].str.contains('Britain') |
            medals['country'].str.contains('Russia') | medals['country'].str.contains('Soviet'), 'continent'] = 'EU'
medals.loc[medals.country.isin(['Cuba', 'Jamaica']), 'continent'] = 'SA'
medals['continent'] = medals['continent'].fillna('OT')
medals
```

#### 9. ตั้งให้ ชื่อทวีปเป็น index ชั้นนอกและชื่อประเทศเป็น index ชั้นใน

```python
medals.set_index(['continent', 'country'], inplace=True)
medals
```

#### 10.	เรียงลำดับชื่อทวีปจากมากไปน้อยและในทวีปเรียงลำดับชื่อประเทศจากน้อยไปมาก

```python
medals.sort_values(['continent', 'country'], ascending=[False, True], inplace=True)
medals
```

#### 11.	จัดคอลัมน์ให้เป็นสองระดับ โดยระดับแรกคือฤดู และระดับที่สองคือจำนวนเกมส์และเหรียญที่ได้ (ดู ตัวอย่างจากเอกสารการสอน) ดังนั้นตอนนี้ข้อมูลจะมีสองระดับทั้ง index (ทวีปและประเทศ) และ column (ฤดูและจำนวนเกมและเหรียญ)

```python
mul = pd.MultiIndex.from_tuples([('Summer', 'Game'), ('Summer', 'Gold'), ('Summer', 'Sliver'), ('Summer', 'Bronze'),
                                 ('Winter', 'Game'), ('Winter', 'Gold'), ('Winter', 'Sliver'), ('Winter', 'Bronze')])
mul
```

```python
old_col = medals.columns
medals.columns = mul
new_medals = medals.copy()
new_medals
```

#### 12.	ย้ายระดับของฤดูมาเป็น index ชั้นในสุด

```python
newer_medals = new_medals.stack(0).fillna(0)
newer_medals.index.names = ['Continent', 'Country', 'Season']
newer_medals
```

#### 13.	ปรับตารางให้เป็นแบบ long โดยหนึ่งแถวจะมีคอลัมน์คือ ทวีป ประเทศ ฤดูแข่งขัน จำนวนเกม ประเภทเหรียญ จำนวนเหรียญ
#### ตั้งแต่นี้เป็นต้นไปสามารถใช้การคำนวณต่อจากตารางแบบ wide หรือ long ที่ได้ทำไว้แล้ว

```python
test = pd.melt(newer_medals.reset_index(),
               id_vars=['Continent', 'Country', 'Season', 'Game'],
               var_name='Medal',
               value_name='Count')
test.sort_values(by=['Continent', 'Country', 'Medal'])
# Count = จำนวนเหรียญ
```

```python
test.loc[test.Country == 'Austria']
```

#### 14.	แสดงจำนวนเหรียญทองของโอลิมปิกฤดูร้อนของแต่ละทวีปจากมากไปน้อย

```python
g_medals = test.loc[(test['Season'] == 'Summer') & (test['Medal'] == 'Gold')]
g_medals
```

```python
sort_g_medals = g_medals.groupby(g_medals.Continent).sum()
sort_g_medals
```

```python
sort_g_medals.reset_index(inplace=True)
sort_g_medals.sort_values(by=['Count'], ascending=False)
```

#### 15.	นับว่าแต่ละทวีปมีกี่ประเทศ (ลองใช้คำสั่ง size() ต่อจากการ groupby() )

```python
test.groupby(test.Continent, axis=0).size()
```

```python
test.groupby(test.Continent, axis=0).count()
```

```python
len( test.groupby(test.Continent, axis=0) )
```

#### 16.	แสดงตารางรวมเหรียญฤดูร้อนแต่ละประเภทของทุกประเทศในทวีปเดียวกัน แสดงผลแยกตามทวีป

```python
summer_medals = test.loc[(test['Season'] == 'Summer'), ['Continent', 'Country', 'Count']]
sum_summer_medals = summer_medals.groupby(['Continent','Country']).sum()
sum_summer_medals
```

#### 17.	แสดงตารางสรุปรวมเหรียญแต่ละประเภทของทุกประเทศแยกตามฤดูแข่งขัน

```python
test.loc[:, ['Country', 'Season', 'Medal', 'Count']].groupby(['Season','Country','Medal']).sum()
```

#### 18.	หาค่าเฉลี่ยของจำนวนเหรียญทองฤดูร้อนต่อประเทศที่เข้าร่วมแข่งขัน แยกตามทวีป

```python
g_medals.groupby(g_medals.Continent, axis=0).agg(AVG_Gold=('Count', 'mean'))
```

#### 19.	หาจำนวนประเทศที่เคยเข้าร่วมการแข่งขันโอลิมปิค แยกตามทวีปและฤดูแข่งขัน แสดงในตารางเดียวกัน

```python
summer_oly = test.loc[(test['Game'] > 0) & (test['Season'] == 'Summer'), ['Continent', 'Country']]
summer_oly.drop_duplicates(inplace=True)

winter_oly = test.loc[(test['Game'] > 0) & (test['Season'] == 'Winter'), ['Continent', 'Country']]
winter_oly.drop_duplicates(inplace=True)

s = summer_oly.groupby('Continent').count()
w = winter_oly.groupby('Continent').count()
```

```python
s
```

```python
# oly = pd.merge(s, w, left_on='Continent', right_on='Continent')
oly = pd.merge(s, w, on='Continent')
oly.columns = ['Summer', 'Winter']
oly
```

#### 20.	หาค่าเฉลี่ยของจำนวนเหรียญทองต่อประเทศที่เข้าร่วมแข่งขัน แยกตามทวีปและฤดูแข่งขัน

```python
All_g_medals = test.loc[(test['Medal'] == 'Gold') & (test['Game'] > 0)]
All_g_medals.groupby(['Continent', 'Season']).agg(AVG_GOLD = ('Count','mean'))
```
