#### 28. ให้สร้างคอลัมน์ใหม่ในตาราง medals ให้ชื่อว่า score โดยคำนวณคะแนนรวมจากเหรียญที่แต่ละประเทศได้ (ทั้งฤดูร้อนและหนาว) โดยให้เหรียญทองได้ 3 คะแนน เหรียญเงินได้ 2 คะแนน เหรียญทองแดงได้ 1 คะแนน แล้วแสดงข้อมูล 10 ประเทศแรกที่ได้คะแนนเยอะที่สุด

```python
medals
```

```python
#sum of season
all_gold = medals[['gold_summer', 'gold_winter']].sum(axis=1)
all_silver = medals[['silver_summer', 'silver_winter']].sum(axis=1)
all_bonz = medals[['bonz_summer', 'bonz_winter']].sum(axis=1)
```

```python
all_gold
```

```python
all_silver
```

```python
all_bonz
```

```python
score = (all_gold * 3) + (all_silver * 2) + (all_bonz * 1)
score
```

```python
medals['score'] = score
medals
```

#### 29. การเปรียบเทียบคะแนนจากเหรียญทั้งหมดอาจจะไม่แฟร์ เพราะแต่ละประเทศเข้าร่วมการแข่งขันไม่เท่ากัน  ดังนั้นให้สร้างคอลัมน์ใหม่ในตาราง medals ชื่อว่า perGame โดยคำนวณคะแนนต่อการเข้าร่วมแข่งขัน
#### (คะแนนคำนวณเหมือนเดิม แต่หารด้วยจำนวนเกมที่เข้าแข่งขัน แยกกันระหว่างฤดูร้อนกับฤดูหนาว ค่า perGame จะเป็นผลบวกของคะแนนของฤดูร้อนกับฤดูหนาว) แสดงข้อมูล 10 ประเทศที่ได้คะแนนน้อยที่สุด
##### 1. ระวัง ตรวจสอบผลลัพธ์ให้ดี มี surprise นะข้อนี้

```python
score_summer = (medals['gold_summer']*3) + (medals['silver_summer']*2) + (medals['bonz_summer']*1)
score_winter = (medals['gold_winter']*3) + (medals['silver_winter']*2) + (medals['bonz_winter']*1)
```

```python
avg_summer = score_summer/medals['summer']
avg_winter = score_winter/medals['winter']
```

```python
perGame = avg_summer.fillna(0) + avg_winter.fillna(0)  # fillna(0) : change NaN to 0
perGame
```

```python
medals['perGame'] = perGame
medals.sort_values('perGame').head(10)  #head(n) = show n values
```
