#### 19.	โหลดไฟล์ชื่อ King_County_Sheriﬀ_s_Oﬃce.csv ซึ่งเก็บการปฏิบัติงานของเจ้าหน้าที่ตำรวจไว้ จงสร้าง ตารางแสดงจำนวนอาชญากรรมผิดกฎหมายแยกตามประเภทของการทำผิดกฎหมาย (ในคอลัมน์ incident_parent_type) และตามรหัสไปรษณีย์ (zip) จากนั้นแสดงตารางนี้

```python
kcsOfice = pd.read_csv("King_County_Sheriff_s_Office.csv")
kcsOfice
```

```python
gType = kcsOfice.groupby(["parent_incident_type","zip"]).sum()
gType
```

#### 20.	เราสนใจอาชญากรรมที่เกี่ยวข้องกับบ้าน ดังนั้นเราจะเน้นที่อาชญากรรมสามประเภทคือ 'Property Crime','Breaking & Entering’ กับ ‘Drugs’ ให้คำนวณตารางใหม่ที่แสดงว่าแต่ละ zip นั้นมีอาชญากร รมทั้งสามเกิดขึ้นคิดเป็นกี่เปอร์เซ็นเทียบกับอาชญากรรมประเภทนั้นรวมในทุก zip และแทนตำแหน่งที่ไม่ มีข้อมูลด้วย 0.0 ด้วย แสดงตารางผลลัพธ์

```python
gType = kcsOfice.groupby(["parent_incident_type","zip"]).count()
byZip = gType.loc[["Property Crime",'Breaking & Entering',"Drugs"]].unstack().incident_id.sum()
byZip
```

```python
byType = gType.loc[["Property Crime",'Breaking & Entering',"Drugs"]].unstack().incident_id.sum(axis=1)
byType
```

```python
byAll = gType.loc[["Property Crime",'Breaking & Entering',"Drugs"]].incident_id
byAll
```

```python
gbyZip3 = kcsOfice.loc[kcsOfice.parent_incident_type.isin(["Property Crime",'Breaking & Entering',"Drugs"])].groupby(["parent_incident_type","zip"]).count()
gbyZip3
```

```python
persentCrime = (gbyZip3.loc[gbyZip3.index.get_level_values('parent_incident_type')].city  * 100 / byType)
persentCrime
```

#### 21.	นำข้อมูลเปอร์เซ็นอาชญากรรมทั้งสามประเภทไปรวมกับตารางการขายบ้าน (ตารางแรก) ที่ตำแหน่งรหัส ไปรษณีย์เดียวกัน แสดงตาราง

```python
persentCrime = persentCrime.reset_index()
persentCrime
```

```python
persentCrime = persentCrime.rename(columns={0 : "persent","zip" : "zipcode"})
persentCrime.zipcode = persentCrime.zipcode.astype(np.int64)
persentCrime
```

```python
house_sales['zipcode']
```

```python
house_crime = pd.merge(house_sales,persentCrime,on="zipcode")
#house_crime.loc[house_crime.zipcode == 98001,"parent_incident_type"]
house_crime
```
