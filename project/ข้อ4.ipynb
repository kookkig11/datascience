 # เมื่อแบ่งช่วงของความพึงพอใจเป็น 5 ช่วง (พอใจมากที่สุด,พอใจค่อนข้างมาก,พอใจปานกลาง,พอใจน้อย,พอใจน้อยที่สุด) ในช่วงของความพึงพอใจใน 5 ช่วงนี้ในแต่ล่ะช่วงนั้นมีความแตกต่างกับช่วงอื่นในด้านใดบ้าง

```python
chefaccept = pd.read_csv("../restaurant-data/chefmozaccepts.csv")
chefcuisine = pd.read_csv("../restaurant-data/chefmozcuisine.csv")
chefhour = pd.read_csv("../restaurant-data/chefmozhours4.csv")
chefparking = pd.read_csv("../restaurant-data/chefmozparking.csv")
geoplace = pd.read_csv("../restaurant-data/geoplaces2.csv")
rating = pd.read_csv("../restaurant-data/rating_final.csv")

usercuisine = pd.read_csv("../restaurant-data/usercuisine.csv")
userpayment = pd.read_csv("../restaurant-data/userpayment.csv")
userprofile = pd.read_csv("../restaurant-data/userprofile.csv")
```

```python
placerating = pd.merge(rating,geoplace,on="placeID")
placerating = pd.merge(placerating,chefaccept,on="placeID")
placerating = pd.merge(placerating,chefcuisine,on="placeID")
placerating = pd.merge(placerating,chefhour,on="placeID")
placerating = pd.merge(placerating,chefparking,on="placeID")
placerating

```

```python
placerating.columns
```

```python
placerating = placerating.loc[:,['userID', 'placeID', 'rating', 'food_rating', 'service_rating',
        'name', 'address', 'city',
       'state', 'country', 'fax', 'zip', 'alcohol', 'smoking_area',
       'dress_code', 'accessibility', 'price', 'url', 'Rambience', 'franchise',
       'area', 'other_services', 'Rpayment', 'Rcuisine', 'hours', 'days',
       'parking_lot']]
placerating
```

mean of rating

```python
avgRating = placerating.groupby("placeID").mean().sort_values("rating",ascending=False)
avgRating.columns = ['avgrating', 'avgfood_rating', 'avgservice_rating']
avgRating = avgRating.reset_index()
avgRating
```

- placerating merge with avg rating

```python
placerating = pd.merge(placerating,avgRating,on="placeID")
placerating
```
