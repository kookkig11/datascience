```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

## Loading data from csv

```python
profile = pd.read_csv('userprofile.csv')
profile.head(5)
```

```python
cuisine = pd.read_csv('usercuisine.csv')
cuisine.head(5)
```

```python
chef_cruise = pd.read_csv('chefmozcuisine.csv')
chef_cruise.head(5)
```

```python
geo = pd.read_csv('geoplaces2.csv')
geo.head(5)
```

```python
rating = pd.read_csv('rating_final.csv')
rating.head(5)
```

## variables of data

|var|data|
|:---|:---|
|`profile`|userprofile.csv|
|`cuisine`|usercuisine.csv|
|`chef_cruise`|chefmozcuisine.csv|
|`geo`|geoplaces2.csv|
|`rating`|rating_final.csv|

```python
all_data = [profile, cuisine, chef_cruise, geo, rating]
all_data_str = ["profile", "cuisine", "chef_cruise", "geo", "rating"]
```

## attributes for each variable

```python
attribute = pd.DataFrame(columns=['var'])
for i in range(len(all_data)):
    attribute.loc[i, 'var'] = all_data_str[i]
    for j in range(len(list(all_data[i].columns))):
        attribute.loc[i, all_data[i].columns[j]] = True
        
attribute.fillna(False, inplace=True)
attribute
```

## intersect columns

```python
col = attribute.columns
out = []
for i in range(1, len(col)):
    if (sum(attribute.loc[:, col[i]]) > 1):
        out.append(col[i])

out.insert(0, 'var')
attribute[out]
```

## add bmi column in profile

```python
profile['bmi'] = profile.weight / profile.height**2
profile.head(5)
```

```python
# cut bmi into string
bmi_str = ['underweight', 'normal', 'overweight', 'obese', 'extreamly obese']

profile['bmi_str'] = pd.cut(profile.bmi, 
       [0, 18.5, 23, 25, 30, np.inf],
       labels=bmi_str
)
profile.head(5)
```

## merge data

|var|data1|data2|key|
|:---|:---|:---|:---|
|`total`|profile|rating|`userID`|
|`total`|total|cuisine|`placeID`|

```python
total = profile.loc[:, ['userID', 'bmi_str']].merge(rating, how='right')

total = total.merge(cuisine, how='outer').rename(columns={'Rcuisine':'res_type'})

# summary = pd.crosstab(index=[total.bmi_str, 
#                    total.userID, total.res_type], 
#             columns=total.rating, 
#             values=total.rating, 
#             aggfunc='count').fillna(0)

# reshape pivot to get value of rating
summary = total.pivot_table(index=['bmi_str', 'userID', 'placeID'], columns='rating', aggfunc='count')['food_rating']

# merge restaurant type and rename
summary = summary.reset_index().merge(cuisine).rename(columns={'Rcuisine':'res_type'})

# slice only necessary then set index
summary = summary.loc[:, ['bmi_str', 'userID', 'res_type', 0, 1, 2]].set_index(['bmi_str', 'userID', 'res_type'])
summary
```

## summary data of rating

```python
# convert data to percent
percent = summary.copy()

for i in bmi_str:
    sum_each = percent.loc[i].sum()
    for j in range(len(percent.loc[i].index)):
        percent.loc[i].iloc[j] = percent.loc[i].iloc[j] / sum_each * 100

percent
```

```python
# display data
for i in bmi_str:
    print('---------------' + i + '----------------')
    count = percent.loc[i].groupby('res_type').count()
    
    for j in range(3):
        print('rating = ' + str(j))
        print(percent.loc[i].groupby('res_type').sum().loc[:, j].nlargest(n=5))
        print()

    print('\n-----------------------------------')
```
